```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as col
import numpy as np
plt.style.use('seaborn')

def plot_policy(i):
    file = f'./kochi/state_ql_mod_30_15_1000/sim_{i:09d}.csv'
    df = pd.read_csv(file,names=['node','d1','d2','d3','d4','d5','d6','d7','d8','d9','d10',
    'q1','q2','q3','q4','q5','q6','q7','q8','q9','q10','f1','f2','f3','f4','f5','f6','f7',
    'f8','f9','f10'])
    a = df.to_numpy()
    # plt.imshow(a[:,11:16],aspect='auto',norm=col.Normalize(),cmap=plt.cm.Pastel1_r)
    plt.pcolormesh(a[:,11:16],cmap=plt.cm.hot_r)
    plt.hlines(4314,0,4,linestyle='--',colors='k')
    plt.ylim(0,6000)
    plt.text(4,5000, f'Policy {i:03d}', bbox={'facecolor': 'white', 'pad': 10})
    plt.colorbar()
    plt.show()
    plt.close()
    return a

for i in range(150,155):
    plot_policy(i)
```

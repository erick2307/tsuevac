```python
plt.style.use("default")
fig, ax = plt.subplots(figsize=(10, 5))
s = 60
c = "k"
for md in [14]:  # meandeparture:
    folder = f"state_db_ql_mod_{s}_{md}"
    df = pd.read_csv(
        f"../arahama/{folder}/survivorsPerSim_1x10000.csv", names=["t", "r"]
    )
    step = 10
    # mean of raw data stepwise
    temp =[df['r'].values[i-step:i].mean() / 2723 for i in range(step,len(df['r'])+1)]
    temp2 = [temp[0]]*(step-1)
    df['rr'] = temp2 + temp
    # std stepwise
    temp =[df['rr'].values[i-step:i].std() for i in range(step,len(df['rr'])+1)]
    temp2 = [temp[0]]*(step-1)
    df['std'] = temp2 + temp
    # upper and lower bounds
    df["upb"] = [x + y for x, y in zip(df["rr"], df["std"])]
    df["lwb"] = [x - y for x, y in zip(df["rr"], df["std"])]
    # plots
    # df['upb'].plot(ax=ax,color=c,marker=None,ls=None,alpha=0.4)
    df["rr"].plot(ax=ax, color=c, marker=None, ls="-", label=f"$\sigma$={md}")
    # df["mean"].plot(ax=ax, color=c, marker=None, ls="-", label=f"$\sigma_{{mean}}$={md}")
    # df['std'].plot(ax=ax,color=c,marker=None,ls='-',label=f'$\sigma$={md}',alpha=0.4)
    # df['lwb'].plot(ax=ax,color=c,marker=None,ls=None,alpha=0.4)
    ax.fill_between(df.index.tolist(), df["upb"], df["lwb"], facecolor=c, alpha=0.4)
plt.legend()
plt.hlines(1,0,10000,ls='--',color='k',alpha=0.5)
plt.ylim(0,1.1)
plt.xlim(0,10000);
```

## Lets find the best policy in the last 10 epochs

```python
best_policies = {}
for s in [15, 30, 60]:
    best_policies[str(s)] = []
    for md in [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60]:  # meandeparture:
        folder = f"state_db_ql_mod_{s}_{md}"
        df = pd.read_csv(
            f"../arahama/{folder}/survivorsPerSim_1x100.csv", names=["t", "r"]
        )
        best_policies[str(s)].append(df["r"][-10:].idxmax())

```

```python
rows = list(best_policies)
cell_text = list(best_policies.values())
columns = [str(a) for a in [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60]]

```

```python
fig, axs = plt.subplots(1, 1, figsize=(8, 4))
axs.axis("tight")
axs.axis("off")
plt.table(cellText=cell_text, rowLabels=rows, colLabels=columns, loc="center")
plt.title("Best Policies epoch")

```

```python
new_dfs = {}
for i, df in enumerate(dfs.values()):
    new_dfs[i] = df.loc[~(df==0).all(axis=1)]
```
